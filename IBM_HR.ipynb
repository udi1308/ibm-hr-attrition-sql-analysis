# Key Business Insight

1. Overtime Strongly Impacts Attrition

Employees working overtime showed significantly higher attrition compared to employees who did not work overtime. This indicates that excessive workload and employee burnout may be major reasons behind employees leaving the organization.

2. Sales Department Has the Highest Attrition

The Sales department experienced the highest employee turnover among all departments. High pressure, targets, and workload may contribute to increased attrition in this department.

3. Lower Salary Employees Are More Likely to Leave

Employees with lower monthly income were found to have higher attrition rates. Compensation and salary satisfaction appear to play an important role in employee retention.

4. Younger Employees Show Higher Turnover

Employees in the age group of 25–35 showed higher attrition compared to older employees. Younger professionals may be seeking faster career growth and better opportunities.

5. Poor Work-Life Balance Affects Retention

Employees with lower work-life balance ratings were more likely to leave the company. This suggests that maintaining a healthy work-life balance is important for employee satisfaction and retention.

6. Job Satisfaction Directly Influences Attrition

Employees with low job satisfaction levels showed a higher tendency to leave the organization. Employee engagement and workplace satisfaction are critical factors in reducing attrition.

7. Employees with Shorter Tenure Leave More Frequently

Most attrition was observed among employees who had spent fewer years at the company. This indicates possible issues in onboarding, training, or early employee engagement.

8. Certain Job Roles Have Higher Attrition

Sales Executives and Laboratory Technicians showed comparatively higher attrition rates than other job roles. Role-specific stress and limited growth opportunities may be contributing factors.

9. Long Distance from Home Slightly Increases Attrition

Employees living farther from the workplace showed slightly higher attrition trends. Long commuting time may negatively affect employee satisfaction and productivity.

10. Single Employees Have Higher Attrition

Single employees showed higher attrition compared to married employees. Younger and single employees may be more likely to switch jobs for better opportunities and career advancement.
